```python
# Set token (private)
ACCESS_TOKEN = ''
```

```python
#*** REST API
# Code from https://developers.zenodo.org/#introduction

import requests

r = requests.get("https://zenodo.org/api/deposit/depositions", params={'access_token': ACCESS_TOKEN})
r.status_code
# r.json()
```

```python
# Setup empty upload
headers = {"Content-Type": "application/json"}
r = requests.post('https://zenodo.org/api/deposit/depositions',
                   params={'access_token': ACCESS_TOKEN}, json={},
                   headers=headers)
```

```python
r.status_code
```

```python
r.json()
```

```python
# Access to json() objects as dict
r.json()['metadata']['prereserve_doi']['doi']
```

```python
# Upload a file

# Get the deposition id from the previous response
deposition_id = r.json()['id']
# Set file
data = {'name': 'zenodo_test.py'}
files = {'file': open('zenodo_test.py', 'rb')}

# Upload
r = requests.post('https://zenodo.org/api/deposit/depositions/%s/files' % deposition_id,
                   params={'access_token': ACCESS_TOKEN}, data=data,
                   files=files)
r.status_code
```

```python
r.json()
```
