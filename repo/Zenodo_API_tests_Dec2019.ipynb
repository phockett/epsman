```python
#**** Remove all items from Zendod (unpublished only)
for item in response.json():
    print(item['id'])
    r = requests.delete('https://zenodo.org/api/deposit/depositions/%s' % item['id'],
                    params={'access_token': ACCESS_TOKEN})
    print(r.status_code)
    
    if r.status_code == 204:
        print('***Item removed')
```

```python
#*** Remove files from repo (unpublished only)

for item in response.json():
    print(item['id'])
    if 'files' in item.keys():
        for fileItem in item['files']:
            r = requests.delete(f"https://zenodo.org/api/deposit/depositions/{item['id']}/files/{fileItem['id']}",
                            params={'access_token': ACCESS_TOKEN})
            print(r.status_code)

            if r.status_code == 204:
                print('***Item removed')

#         print(f"https://zenodo.org/api/deposit/depositions/{item['id']}/files/{fileItem['id']}")
```
